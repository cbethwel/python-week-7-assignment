```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
try:
    df = pd.read_csv('iris.csv')
    print("Dataset loaded successfully.")
except FileNotFoundError:
    print("File not found. Please check your path.")
```

```python
df.head()
```

```python
df.info()
```

```python
print(df.isnull().sum())

# If there are missing values:
df = df.dropna()  # OR you can fill using df.fillna(method='ffill')
```

```python
df.describe()
```

```python
grouped = df.groupby('variety').mean()
print(grouped)
```

```python
plt.figure(figsize=(10,5))
plt.plot(df.index, df['petal.length'], label='Petal Length')
plt.title('Petal Length across Samples')
plt.xlabel('Sample Index')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.show()
```

```python
species_means = df.groupby('variety')['petal.length'].mean()
species_means.plot(kind='bar', color=['orange', 'green', 'blue'])
plt.title('Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length (cm)')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['sepal.length'], bins=20, color='purple')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='sepal.length', y='petal.length', hue='variety', data=df)
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.show()
```
